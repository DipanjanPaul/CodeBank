```python
from datetime import date, time
import xlsxwriter
import pandas as pd

import re

from openpyxl import load_workbook, Workbook
from openpyxl.utils import get_column_letter
from openpyxl.worksheet.datavalidation import DataValidation

from sqlalchemy import create_engine
from sqlalchemy.sql import text

from copy import copy
```

```python
template = "../data/external/Bid Template SKUs Meta Data - DP.xlsx"
output = "../data/interim/Formatted_Bid.xlsx"
```

```python
##sheet_name='Std Names + Takeoff Quantities'
template_wb = load_workbook(filename = template)
template_ws = template_wb['Std Names + Takeoff Quantities']
```

```python
engine = create_engine('postgresql://localhost/TB')
conn = engine.connect()  
```

```python
def get_sku_vald_list(category, sub_category, work_type):
    s =  "SELECT * \
      FROM sku_data \
      WHERE category = %(category)s \
        AND sub_category = %(sub_category)s \
        AND work_type = %(work_type)s"
    
    df = pd.read_sql(s, conn, params={'category': category, 'sub_category': sub_category, 
                                      'work_type' : work_type})
    
    ##return(','.join(df.rfp_spec.str.replace(', ', '/')))
    ##return(','.join(df.rfp_spec))
    return(','.join(df.rfp_spec.str.replace(', ', '+').replace('++', '+')))

```

```python
output_wb = Workbook()
output_ws = output_wb.create_sheet(template_ws.title)

dv = {}

m_cell_st = []
m_cell_end = []

cost_center_prev = ""
category_prev = ""

for row in template_ws:
    
    cost_center = row[1].value
    
    if (re.sub('Section [A-Z] ', '', str(row[0].value)).upper() != 'NONE' ):
        category = re.sub('Section [A-Z] ', '', str(row[0].value)).upper()
        
    sub_category = str(row[2].value).upper()
    work_type = str(row[3].value).upper()
    
    print(category, sub_category, work_type)
    
    if ((re.match("[0-9]{1}-[0-9]{2}-[0-9]{4}", str(cost_center_prev)) is None) and \
        (re.match("[0-9]{1}-[0-9]{2}-[0-9]{4}", str(cost_center)) is not None)):
        m_cell_st.append(row[1].coordinate)
        
    if ((re.match("[0-9]{1}-[0-9]{2}-[0-9]{4}", str(cost_center_prev)) is not None) and \
        (re.match("[0-9]{1}-[0-9]{2}-[0-9]{4}", str(cost_center)) is None)):
        m_cell_end.append(row[1].coordinate)
        
    
    for cell in row:
        
        if  (cell.coordinate[0] not in ['A', 'B', 'C', 'D']) \
        and (re.match("[0-9]{1}-[0-9]{2}-[0-9]{4}", str(cost_center))):
            
            if cell.coordinate[0] == 'G':
                output_ws[cell.coordinate].value = None
                ##x = output_ws[cell.coordinate]
                
                c_val = cell.coordinate
                drop_down_val = "\"" + "Exclude" + ',' + \
                               get_sku_vald_list(category, sub_category, work_type) + "\""
                
                dv[cell.coordinate] = DataValidation(type="list", promptTitle='Select from list',
                                                     formula1=str(drop_down_val), 
                                                     showDropDown=True, allow_blank=True)
                
                output_ws.add_data_validation(dv[cell.coordinate])
                
                dv[cell.coordinate].add(output_ws[cell.coordinate])
            else:
                output_ws[cell.coordinate].value = None
                
        else:
            output_ws[cell.coordinate].value = cell.value
        
        if cell.has_style:
            output_ws[cell.coordinate].font = copy(cell.font)
            output_ws[cell.coordinate].border = copy(cell.border)
            output_ws[cell.coordinate].fill = copy(cell.fill)
            output_ws[cell.coordinate].number_format = copy(cell.number_format)
            output_ws[cell.coordinate].protection = copy(cell.protection)
            output_ws[cell.coordinate].alignment = copy(cell.alignment)
            
        cost_center_prev = cost_center
            
output_ws.merge_cells('E2:G2')

m_cell_st = ['A' + x[1:] for x in m_cell_st]
m_cell_end = ['A' + str(int(x[1:]) - 1 )for x in m_cell_end]
m_ranges = [x+':'+y for x,y in zip(m_cell_st,m_cell_end)]

for itm in m_ranges:
    output_ws.merge_cells(itm)
    
for i in range(template_ws.max_column):
    output_ws.column_dimensions[get_column_letter(i+1)].width = \
                template_ws.column_dimensions[get_column_letter(i+1)].width
```

```python
del output_wb['Sheet']
output_wb.save(output)
output_wb.close()
```
