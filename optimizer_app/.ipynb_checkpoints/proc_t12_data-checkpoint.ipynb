```python
def process_yearly_file(infile, worksheet):
    #Part 1
    df = pd.read_excel(infile, sheet_name=worksheet, header=None)
    df.columns = ["Col_" + str(x) for x in df.columns]
    df.drop(columns='Col_14', inplace=True)
    df = df[-df['Col_0'].isnull()]
    
    x = df.apply(lambda x: np.sum(x.isnull()), axis=0)
    df.drop(columns=x[x>150].index, inplace=True)

    hdr = df.iloc[0]

    df = df.iloc[1:, :]

    df["chg_cd_1"] = df["Col_0"].apply(lambda x: x.split('-')[0])
    df["chg_cd_2"] = df["Col_0"].apply(lambda x: x.split('-')[1])

    df = df[df.chg_cd_2 != '9999']

    df = df[-df.chg_cd_1.isin(['4000', '6000'])]

    df.reset_index(drop=True, inplace=True)  
    
    #Part 2
    #print(df.head())
    catg_map = df[df['Col_2'].isnull()][['Col_0', 'Col_1']]
    catg_map.columns = ["chg_cd_composite", "chg_desc"]
    catg_map["chg_cd_1"] = catg_map.chg_cd_composite.apply(lambda x: x.split('-')[0])
    catg_map["chg_cd_2"] = catg_map.chg_cd_composite.apply(lambda x: x.split('-')[1])
    catg_map['i_start'] = catg_map.index

    i_end = list(catg_map.index[1:].values)
    i_end.append(df.shape[0])

    catg_map['i_end'] = i_end

    catg_map['repeat_cnt'] = catg_map.i_end - catg_map.i_start
    catg_map.reset_index(drop=True, inplace=True)
    
    x = catg_map.apply(lambda x: list(np.repeat(x[1].strip(), x[6])), axis=1)
    catg = [itm for sublist in x for itm in sublist]
    
    df['Category'] = catg
    
    #Part 3
    df = df[-df.iloc[:,2:14].apply(lambda x: np.sum(x.isnull()) == 12, axis=1)]
    df.reset_index(drop=True, inplace=True)
    if ('Col_0' in df.columns):
        df.drop(columns='Col_0', inplace=True)
    
    df.columns = ['SubCategory', '01_JAN', '02_FEB', '03_MAR', '04_APR', '05_MAY', '06_JUN', '07_JUL',
                  '08_AUG', '09_SEP', '10_OCT', '11_NOV', '12_DEC', 'CHG_CD1', 'CHG_CD2', 'Category']

    df = df[['Category', 'SubCategory', '01_JAN', '02_FEB', '03_MAR', '04_APR', '05_MAY', '06_JUN', '07_JUL',
                  '08_AUG', '09_SEP', '10_OCT', '11_NOV', '12_DEC', 'CHG_CD1', 'CHG_CD2']]
    
    df.SubCategory = [x.strip() for x in df.SubCategory]
    
    #Part 4
    df = df.melt(id_vars=['Category', 'SubCategory', 'CHG_CD1', 'CHG_CD2'],
                  value_vars=['01_JAN', '02_FEB', '03_MAR', '04_APR', '05_MAY', '06_JUN', 
                              '07_JUL', '08_AUG', '09_SEP', '10_OCT', '11_NOV'])
    
    df.columns = ['Category', 'SubCategory', 'CHG_CD1', 'CHG_CD2', 'MM', 'Value']
    df['P_Year'] = worksheet
    return(df)
```

```python
infile = "/Users/dipanjanpaul/Winpra/Analysis/(GK) Sienna Heights - T12 2015-2016-2017.xlsx"
wb_sheets = pd.ExcelFile(infile).sheet_names
wb_sheets
#worksheet = 'T12-2015'
```

```python
final = []
for itm in wb_sheets[:2]:
    print(itm)
    final.append(process_yearly_file(infile, itm))
    
```

```python
final_df = pd.concat(final, axis=0)
```

```python
final_df.to_csv("/Users/dipanjanpaul/Winpra/dash/data/mkt_analysis.csv", header=True, index=False)
```
